```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.cluster import SpectralClustering
```

```python
clustering_model = SpectralClustering(n_clusters=6, affinity='nearest_neighbors', n_neighbors=12)
df1['cluster'] = clustering_model.fit_predict(df1)

# Ã‰tape 3: CrÃ©er la visualisation interactive avec Plotly
fig = px.scatter_3d(df1, x='x', y='y', z='z', color='cluster', symbol='cluster', size_max=1)
fig.show()
```

```python
from sklearn.linear_model import LinearRegression
X = df1[['x']]
y = df1['y']

regression_model = LinearRegression()
regression_model.fit(X, y)

# Obtenir les coefficients de la droite (a et b) de l'Ã©quation y = ax + b
a = regression_model.coef_[0]
b = regression_model.intercept_
```

```python

# Ã‰tape 3: CrÃ©er une visualisation en 2D pour afficher les points et la droite moyenne ajustÃ©e
fig = go.Figure()

# Ajouter les points
fig.add_trace(go.Scatter(
    x=df1['x'],
    y=df1['y'],
    mode='markers',
    marker=dict(size=5, color='blue')
))

# CrÃ©er la droite moyenne Ã  partir des coefficients trouvÃ©s
x_line = np.linspace(df1['x'].min(), df1['x'].max(), 100)
y_line = a * x_line + b

# Ajouter la droite moyenne Ã  la figure
fig.add_trace(go.Scatter(x=x_line, y=y_line, mode='lines', line=dict(color='red')))

# Afficher la figure
fig.show()
```
